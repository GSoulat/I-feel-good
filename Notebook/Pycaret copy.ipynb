```python
from pycaret.nlp import *
from pycaret.classification import *

import pandas as pd
```

# Importation du fichier kaggle

```python
df_train_kaggle = pd.read_csv('../data/kaggle.csv')
df_train_kaggle.info()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
tf_vectorizer = CountVectorizer(min_df=.015, max_df=.8,strip_accents='ascii', lowercase=True, stop_words='english' , ngram_range=[1, 3])
```

```python
%time features = tf_vectorizer.fit_transform(df_train_kaggle.Text)
```

```python
features_df = pd.DataFrame(features.toarray(), columns=tf_vectorizer.get_feature_names())
```

```python
df = pd.concat([features_df,df_train_kaggle.Emotion],axis=1)
```

```python
#Shuffle your dataset 
shuffle_df = df.sample(frac=1)

# Define a size for your train set 
train_size = int(0.9 * len(df))

# Split your dataset 
train_df = shuffle_df[:train_size]
test_df  = shuffle_df[train_size:]
```

```python
numerical_features = list(features_df.columns)
%time temp = setup(data = train_df, target = 'Emotion',numeric_features=numerical_features, use_gpu=True)
```

```python
%time lightgbm = create_model('lightgbm')
```

```python
best = compare_models()
```

```python
evaluate_model(best)
```
